# IRIS FLOWER CLASSIFICATION

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df = pd.read_csv('IRIS.csv')
df
```

```python
species_mean = df.groupby('species').agg({
    'sepal_length': np.mean,
    'sepal_width': np.mean,
    'petal_length': np.mean,
    'petal_width': np.mean
})
print(species_mean)
```

```python
features = df.drop("species", axis=1)
labels = df["species"]
labels
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn
```

```python
knn.fit(features, labels)
```

```python
predictions = knn.predict(features)
predictions
```

```python
accuracy = knn.score(features, labels)
accuracy
```

```python
print("Accuracy:", accuracy)
```

```python
# Load the Iris dataset
iris = load_iris()

# Fit the model on the dataset
knn.fit(iris.data, iris.target)

# Create a new flower for prediction
new_flower = [[5.1, 3.5, 1.4, 0.2]]

# Predict the species using the trained model
predicted_species_index = knn.predict(new_flower)

# Map the predicted species index to the species label
predicted_species_label = iris.target_names[predicted_species_index][0]

# Print the predicted species label
print(f"Predicted Species: {predicted_species_label}")
```
